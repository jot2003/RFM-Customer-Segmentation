```python
import pandas as pd
```

```python

file_path = 'data.csv'
data = pd.read_csv(file_path)

data.head()
```

```python
data.info()
```

```python
data.sample(5)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Select the RFM columns for clustering
rfm_data = data[['recency', 'frequency', 'monetary']]

# Standardize the data
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_data)

print(rfm_data)

# Apply K-Means clustering to divide the data into 5 clusters
kmeans = KMeans(n_clusters=5, random_state=42)
data['cluster'] = kmeans.fit_predict(rfm_scaled)

# Map the clusters to the predefined customer segments
cluster_labels = {
    0: 'Khách hàng VIP',
    1: 'Khách hàng Tiềm năng',
    2: 'Khách hàng Thường xuyên',
    3: 'Khách hàng Ngủ đông',
    4: 'Khách hàng Mới'
}

# Assign the cluster labels to the data
data['segment'] = data['cluster'].map(cluster_labels)

# Display the first few rows of the resulting data with segments
data[['user_pseudo_id', 'recency', 'frequency', 'monetary', 'segment']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the plot size and style
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")

# Count the number of customers in each segment
segment_counts = data['segment'].value_counts()
print(segment_counts)

# Create a bar plot for the customer segments
sns.barplot(x=segment_counts.index, y=segment_counts.values, palette='viridis')

# Add labels and title
plt.xlabel('Customer Segments')
plt.ylabel('Number of Customers')
plt.title('Customer Segmentation Based on RFM Model')
plt.xticks(rotation=45)
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import matplotlib.pyplot as plt

# Hierarchical Clustering using Ward's method
Z = linkage(rfm_scaled, method='ward')

# Plotting the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(Z, truncate_mode='lastp', p=12, leaf_rotation=45., leaf_font_size=10., show_contracted=True)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Cluster size')
plt.ylabel('Distance')
plt.show()

# Apply hierarchical clustering to form 5 clusters
hierarchical_clusters = fcluster(Z, 5, criterion='maxclust')

# Map the clusters to the predefined customer segments
data['hierarchical_cluster'] = hierarchical_clusters

# Display the first few rows with the new cluster assignment
data[['user_pseudo_id', 'recency', 'frequency', 'monetary', 'hierarchical_cluster']].head()
```

```python
# Cell 1: Import thư viện cần thiết
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cell 2: Load dữ liệu và hiển thị thông tin cơ bản
file_path = 'data.csv'
data = pd.read_csv(file_path)
print(data.head())
print(data.info())
```

```python
# Cell 3: Chọn các cột RFM và chuẩn hóa dữ liệu
rfm_data = data[['recency', 'frequency', 'monetary']]
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_data)
print(pd.DataFrame(rfm_scaled, columns=['recency', 'frequency', 'monetary']).head())
```

```python
# Cell 4: Tìm số lượng cụm tối ưu bằng phương pháp Elbow
sse = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(rfm_scaled)
    sse.append(kmeans.inertia_)
```

```python
# Cell 5: Vẽ biểu đồ Elbow
plt.figure(figsize=(8, 4))
plt.plot(k_range, sse, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Sum of Squared Distances')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
# Cell 6: Thực hiện phân cụm với k = 5
k = 5
kmeans = KMeans(n_clusters=k, random_state=42)
data['cluster'] = kmeans.fit_predict(rfm_scaled)
print(data['cluster'].value_counts())
```

```python
# Cập nhật gán nhãn khách hàng dựa trên phân tích RFM mới
# Khách hàng VIP: R thấp hoặc trung bình, F cao hoặc trung bình, M cao
# Khách hàng Thường xuyên: R cao, F trung bình hoặc thấp, M trung bình hoặc thấp
# Khách hàng Tiềm năng: R trung bình, F cao, M trung bình
# Khách hàng Không hoạt động: R rất cao, F thấp hoặc không có, M thấp hoặc không có
# Khách hàng Mới: R thấp, F thấp hoặc trung bình, M trung bình hoặc cao

cluster_labels = {
    0: 'Khách hàng Không hoạt động',  # Cluster 0 có recency rất cao, frequency thấp, monetary thấp
    1: 'Khách hàng Mới',  # Cluster 1 có recency thấp, frequency thấp hoặc trung bình, monetary trung bình hoặc cao
    2: 'Khách hàng Thường xuyên',  # Cluster 2 có recency cao, frequency trung bình hoặc thấp, monetary trung bình hoặc thấp
    3: 'Khách hàng Tiềm năng',  # Cluster 3 có recency trung bình, frequency cao, monetary trung bình
    4: 'Khách hàng VIP'  # Cluster 4 có recency thấp hoặc trung bình, frequency cao hoặc trung bình, monetary cao
}

# Áp dụng gán nhãn cho các cụm dựa trên phân tích trên
data['segment'] = data['cluster'].map(cluster_labels)
```

```python
# Cell 8: Đếm số lượng khách hàng trong từng phân đoạn và tạo biểu đồ
segment_counts = data['segment'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=segment_counts.index, y=segment_counts.values, palette='viridis')

plt.xlabel('Customer Segments')
plt.ylabel('Number of Customers')
plt.title('Customer Segmentation Based on RFM Model')
plt.xticks(rotation=45)
plt.show()
```

```python
# Cell 9: Kiểm tra chi tiết về các giá trị RFM trong mỗi phân đoạn
for segment, subset in data.groupby('segment'):
    print(f"Segment: {segment}")
    print(subset[['recency', 'frequency', 'monetary']].describe())
    print("\n")
```
